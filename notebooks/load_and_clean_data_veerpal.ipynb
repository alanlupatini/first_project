```python
df_cleaned = df.copy()
for col in ['District', 'Location']:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
# clean string columns
str_cols = ['Location']
for col in str_cols:
    if col in df.columns:
        df_cleaned[col] = df_cleaned[col].astype(str).str.strip()
        df_cleaned[col] = df_cleaned[col].str.replace(r'\s+', ' ', regex=True)
        
```

```python

#Remove special characters
df_cleaned['Location'] = df_cleaned['Location'].str.replace('/', ',')
df_cleaned = df_cleaned.assign(Location=df_cleaned['Location'].str.split(',')).explode('Location')
df_cleaned['Location'] = df_cleaned['Location'].str.strip()
```

```python
df_cleaned = df_cleaned[~df_cleaned["Location"].str.contains("Nicht Zuzuordnen")]
```
