```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the CSV
df = pd.read_csv("../data/analysis_tables/most common crimes in alexanderplatz in 2019.csv")

# Preview the data
print(df.head())
print(df.info())
```

```python
crime_category = (
    df.groupby("crime_type")["total_count"]
      .mean()
      .sort_values(ascending=False).head(10)
)

plt.figure(figsize=(10,5))

crime_category.plot(kind="bar")
plt.title("Most common crimes in alexanderplatz in 2019")
plt.xlabel("Crime Type")
plt.ylabel("Total Crimes")

# Save the plot as an image
plt.savefig("../figures/top_10_crimes.png", dpi=300, bbox_inches='tight')

plt.show()
```
