```python
df = df.copy()
for col in ['District', 'Location']:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
 # delete Unnamed: 0
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# clean string columns
str_cols = ['Location']
for col in str_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].str.replace(r'\s+', ' ', regex=True)
        
```

```python
df['Location']
```

```python

#Remove special characters
df['Location'] = df['Location'].str.replace('/', ',')
df = df.assign(Location=df['Location'].str.split(',')).explode('Location')
df['Location'] = df['Location'].str.strip()

# 1. remove noise labels
df['Location'] = df['Location'].replace(r'Bezirk.*Nicht Zuzuordnen', None, regex=True)

# 2. remove street names
df = df[~df['Location'].str.contains(r'\bstr\.', case=False, na=False)]

# 3. normalize direction suffixes
direction_map = {
    r'\bNord\b': 'North',
    r'\bSüd\b': 'South',
    r'\bOst\b': 'East',
    r'\bWest\b': 'West',
    r'\bZentrum\b': 'Central',
    r'\bMitte\b': 'Central'
}
for pattern, replacement in direction_map.items():
    df['Location'] = df['Location'].str.replace(pattern, replacement, regex=True)

# 4. normalize compound directions
compound_map = {
    r'Nordwest': 'Northwest',
    r'Nordnet': 'Northeast',
    r'Südwest': 'Southwest',
    r'Südost': 'Southeast'
}
for pattern, replacement in compound_map.items():
    df['Location'] = df['Location'].str.replace(pattern, replacement, regex=True)

# 5. standardize formatting
df['Location'] = df['Location'].str.strip().str.title()

```

```python
def clean_german_string_final(series):
    """
    Standardizes location names by converting to lowercase first, 
    then handling umlauts, hyphens, and spaces.
    """

    cleaned_string = series.astype(str)
    cleaned_string = cleaned_string.str.lower()
    cleaned_string = cleaned_string.str.replace('ä', 'ae', regex=False)
    cleaned_string = cleaned_string.str.replace('ö', 'oe', regex=False)
    cleaned_string = cleaned_string.str.replace('ü', 'ue', regex=False)
    cleaned_string = cleaned_string.str.replace('ß', 'ss', regex=False)
    cleaned_string = cleaned_string.str.replace('-', '', regex=False)  
    cleaned_string = cleaned_string.str.replace(' ', '_', regex=False)
    cleaned_string = cleaned_string.str.replace('__+', '_', regex=True)
    cleaned_string = cleaned_string.str.strip('_')
    
    return cleaned_string


df['location'] = clean_german_string_final(df['location'])
df['district'] = clean_german_string_final(df['district'])
```
